```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType, DoubleType
from pyspark.sql.functions import current_timestamp, to_timestamp, concat, col, lit
```

```python
dbutils.widgets.text("container", "raw")
dbutils.widgets.text("catalogo", "catalog_au")
dbutils.widgets.text("esquema", "bronze")
dbutils.widgets.text("storageName", "adlsdiegoadama2204")
```

```python
container = dbutils.widgets.get("container")
catalogo = dbutils.widgets.get("catalogo")
esquema = dbutils.widgets.get("esquema")
storageName = dbutils.widgets.get("storageName")

ruta = f"abfss://{container}@{storageName}.dfs.core.windows.net/polizas_example.csv"
```

```python
constructors_schema = StructType(fields=[
    StructField("id_poliza", StringType(), False),
    StructField("id_cliente", StringType(), True),
    StructField("tipo_seguro", StringType(), True),
    StructField("fecha_inicio", DateType(), True),
    StructField("prima_anual", DoubleType(), True),
    StructField("estado", StringType(), True)
])
```

```python
df_constructors = spark.read\
.option('header', True)\
.schema(constructors_schema)\
.csv(ruta)
```

```python
circuits_selected_df = df_constructors.select(col("id_poliza"), 
                                                col("id_cliente"), 
                                                col("tipo_seguro"), 
                                                col("fecha_inicio"), 
                                                col("prima_anual"), 
                                                col("estado"))
```

```python
circuits_selected_df.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.polizas")
```
