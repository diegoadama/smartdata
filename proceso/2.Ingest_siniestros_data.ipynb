```python
from pyspark.sql.types import StructType, StructField, IntegerType, StringType, DateType, DoubleType
from pyspark.sql.functions import current_timestamp, to_timestamp, concat, col, lit
```

```python
dbutils.widgets.text("container", "raw")
dbutils.widgets.text("catalogo", "catalog_au")
dbutils.widgets.text("esquema", "bronze")
dbutils.widgets.text("storageName", "adlsdiegoadama2204")
```

```python
container = dbutils.widgets.get("container")
catalogo = dbutils.widgets.get("catalogo")
esquema = dbutils.widgets.get("esquema")
storageName = dbutils.widgets.get("storageName")

ruta = f"abfss://{container}@{storageName}.dfs.core.windows.net/siniestros_example.csv"
```

```python
races_schema = StructType(fields=[StructField("id_siniestro", StringType(), False),
                                  StructField("id_poliza", StringType(), True),
                                  StructField("fecha_siniestro", DateType(), True),
                                  StructField("monto_reclamado", DoubleType(), True),
                                  StructField("estado_siniestro", StringType(), True)
])
```

```python
races_df = spark.read \
            .option("header", True) \
            .schema(races_schema) \
            .csv(ruta)
```

```python
races_selected_df = races_df.select(col("id_siniestro"), 
                                                col("id_poliza"), 
                                                col("fecha_siniestro"), 
                                                col("monto_reclamado"), 
                                                col("estado_siniestro"))
```

```python
races_selected_df.write.mode("overwrite").insertInto(f"{catalogo}.{esquema}.siniestros")
```
